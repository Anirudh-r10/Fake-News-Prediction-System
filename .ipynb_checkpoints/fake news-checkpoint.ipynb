# FAKE NEWS DETECTION SYSTEM

A fake news detection system employs machine learning techniques, including algorithms like Naive Bayes, logistic regression, and random forest, to automatically identify and categorize misinformation in news articles. This system analyzes various linguistic and contextual features within the content to distinguish between genuine and misleading information. Key components involve natural language processing (NLP) to assess language patterns, source credibility analysis to evaluate historical reliability, and social media monitoring to identify potential misinformation spread. Incorporating user feedback and fact-checking databases further enhances the system's accuracy, contributing to a robust and adaptable approach for combating the proliferation of fake news.

### Cleaning text

```python
def wordopt(text):
    text= text.lower()
    text= re.sub('\[.*?\]','',text)
    text= re.sub("\\W", " ",text)
    text= re.sub('https?://\S+|www\.\S+', '',text)
    text= re.sub('<.*?>+','',text)
    text= re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text= re.sub('\n','',text)
    text= re.sub('\w*\d\w*','',text)
    return text
# cleaning data to convert to lowercases, remove brackets, remove links, Replaces non-word characters with a space,Removes punctuation, Removes newline characters, Removes words containing digits.
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
from tqdm import tqdm 
import re 
import nltk 
nltk.download('punkt') 
nltk.download('stopwords') 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem.porter import PorterStemmer 
from wordcloud import WordCloud
```

```python
from tqdm import tqdm
from nltk.corpus import stopwords

def preprocess_text(text_data): 
    preprocessed_text = [] 
      
    for sentence in tqdm(text_data): 
        preprocessed_text.append(' '.join(token.lower() 
                                  for token in str(sentence).split() 
                                  if token not in stopwords.words('english'))) 
  
    return preprocessed_text
```

- Iterates through each sentence in the input data, tokenizes it into words, converts each word to lowercase, and excludes stopwords.

```python
preprocessed_review = preprocess_text(data['text'].values) 
data['text'] = preprocessed_review
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer 
  
  
def get_top_n_words(corpus, n=None): 
    vec = CountVectorizer().fit(corpus) 
    bag_of_words = vec.transform(corpus) 
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()] 
    words_freq = sorted(words_freq, key=lambda x: x[1], 
                        reverse=True) 
    return words_freq[:n] 
  
  
common_words = get_top_n_words(data['text'], 20) 
df1 = pd.DataFrame(common_words, columns=['Review', 'count']) 
  
df1.groupby('Review').sum()['count'].sort_values(ascending=False).plot( 
    kind='bar', 
    figsize=(10, 6), 
    xlabel="Top Words", 
    ylabel="Count", 
    title="Bar Chart of Top Words Frequency"
) 
# count vectorizing to see the frequency of words
```

```python
x= data['text']
y=data['class']
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

- converting a collection of text documents into a numerical format that can be used as input for machine learning models.

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

LR=LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr =LR.predict(xv_test)
```

```python
from sklearn.metrics import confusion_matrix

# Calculate the confusion matrix
cm_lr = confusion_matrix(y_test, pred_lr)

# Display the confusion matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

- The logistic regression model achieved high precision (99%) for non-fake news and high recall (98%) for fake news. 
- It demonstrated balanced F1-scores (0.99 for non-fake, 0.98 for fake), showcasing a harmonious trade-off between precision and recall.
- The overall accuracy of 98% indicates effective classification of both fake and non-fake news instances.
